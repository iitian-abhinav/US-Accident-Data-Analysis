**Is the distributuion of accidents by hour the same on weekends as on weekdays.**

```python
day_of_week=pd.DataFrame(day_of_week)
day_of_week.rename(columns={'Start_Time':'Day'},inplace=True)
weekends=len(day_of_week[(day_of_week['Day']=='Sunday') | (day_of_week['Day']=='Saturday')])/(24*2)
weekdays=len(day_of_week[(day_of_week['Day']!='Sunday') & (day_of_week['Day']!='Saturday')])/(24*5)
print("Accidents per hour on weekends = ",weekends)
print("Accidents per hour on weekdays = ",weekdays)
```

```python
df_2016=df[df['Start_Time'].dt.year==2016]
df_2017=df[df['Start_Time'].dt.year==2017]
df_2018=df[df['Start_Time'].dt.year==2018]
df_2019=df[df['Start_Time'].dt.year==2019]
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot each year's data in its respective subplot
sns.histplot(df_2016['Start_Time'].dt.month, bins=12, ax=axs[0, 0])
axs[0, 0].set_title('2016')

sns.histplot(df_2017['Start_Time'].dt.month, bins=12, ax=axs[0, 1])
axs[0, 1].set_title('2017')

sns.histplot(df_2018['Start_Time'].dt.month, bins=12, ax=axs[1, 0])
axs[1, 0].set_title('2018')

sns.histplot(df_2019['Start_Time'].dt.month, bins=12, ax=axs[1, 1])
axs[1, 1].set_title('2019')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
df['Source'].unique()
```

**There are 3 sources of data namely MapQuest, MapQuest-Bing and Bing and we will analyse the trend based on both the sources to find the anomaly in the data**

```python
df_2019_bing=df_2019[df_2019['Source']=='Bing']
df_2019_mb=df_2019[df_2019['Source']=='MapQuest-Bing']
df_2019_map=df_2019[df_2019['Source']=='MapQuest']

fig,ax=plt.subplots(1,3)
sns.displot(df_2019_bing['Start_Time'].dt.month,bins=12,ax=ax[0])
ax[0].set_title('Bing Data')
sns.displot(df_2019_mb['Start_Time'].dt.month,bins=12,ax=ax[1])
ax[1].set_title('MapQuest-Bing')
sns.displot(df_2019_map['Start_Time'].dt.month,bins=12,ax=ax[2])
ax[2].set_title('MapQuest')
plt.show()
```
